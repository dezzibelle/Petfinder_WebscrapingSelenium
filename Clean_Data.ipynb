```python
import pandas as pd
import numpy as np
import string
import re
from datetime import datetime
```

```python
import glob
all_files = glob.glob(r'data' + "/*.csv")  #Create list of all csv filenames

with open("combined.csv", 'a') as newcsv:  #Create combined csv of all files
    for file in all_files:
        df = pd.read_csv(file, header = None)
        df.to_csv(newcsv, mode='a', header=False)
```

```python
petsdf = pd.read_csv('combined.csv',header= None)
```

```python
petsdf.columns = ["Index", "Name", "Organization", "Info1", "Info2", "Story", "Script"]
```

```python
petsdf.head()
```

```python
def gender(info):
    if "Male" in info:
        return("Male")
    if "Female" in info:
        return("Female")
    else:
        return("NA")
    
def age(info):
    if "Puppy" in info:
        return("Puppy")
    if "Young" in info:
        return("Young")
    if "Adult" in info:
        return("Adult")
    if "Senior" in info:
        return("Senior")
    else:
        return("NA")
    
def size(info):
    if "Small" in info:
        return("Small")
    if "Medium" in info:
        return("Medium")
    if "Large" in info:
        return("Large")
    if "XL" in info:
        return("XL")
    else:
        return("NA")

    
petsdf['Gender'] = petsdf['Info1'].apply(gender)
petsdf['Age'] = petsdf['Info1'].apply(age)
petsdf['Size'] = petsdf['Info1'].apply(size)

petsdf.head()
```

```python
import json

def findjson(string):
    i = string.find("{\"user_auth\"")
    jtext = string[i:]
    j = jtext.rfind("}(window)")
    jtext = jtext[:j]
    k = jtext.rfind(";")
    jtext = jtext[:k]
    return jtext
```

```python
petsdf['jtext'] = petsdf['Script'].apply(findjson)
petsdf.head()
```

```python
petsdf = petsdf.drop_duplicates(['jtext'])
petsdf = petsdf.loc[petsdf['Name'] != '0']
petsdf.shape
```

```python
def city(jtext):
    try:
        json_object = json.loads(jtext)
        city = json_object["location"]["address"]["city"]
        return city
    except:
        return ''

def state(jtext):
    try:
        json_object = json.loads(jtext)
        state = json_object["location"]["address"]["state"]
        return state
    except:
        return('')

def zipcode(jtext):
    try:
        json_object = json.loads(jtext)
        zipcode = json_object["location"]["address"]["postal_code"]
        return zipcode
    except:
        return('')   

def pubdate(jtext):
    try:
        json_object = json.loads(jtext)
        pubdate = json_object["animal"]["published_at"]
        return pubdate
    except:
        return ''
    
def coat(jtext):
    try:
        json_object = json.loads(jtext)
        coat = json_object["animal"]["coat_length"]
        return coat
    except:
        return('')  
    
petsdf['City'] = petsdf['jtext'].apply(city)
petsdf['State'] = petsdf['jtext'].apply(state)
petsdf['Zip'] = petsdf['jtext'].apply(zipcode)

petsdf['PublishDate'] = petsdf['jtext'].apply(pubdate)
petsdf['Coat'] = petsdf['jtext'].apply(coat)
petsdf.head()
```

```python
def breed(jtext):
    try:
        json_object = json.loads(jtext)
    except:
        return ''
    breed1 = json_object["animal"]["primary_breed"]["name"]
    try:
        breed2 = json_object["animal"]["secondary_breed"]["name"]
    except:
        breed2 = ""
    if json_object["animal"]["is_mixed_breed"] == True:
        mix = "Mix"
    else:
        mix = ""
    return([breed1, breed2, mix])

def goodwith(jtext):
    try:
        json_object = json.loads(jtext)
    except:
        return('')
    if json_object["animal"]["home_environment_attributes"]["good_with_children"] == True:
        kids = "Children"
    else:
        kids = ""
    if json_object["animal"]["home_environment_attributes"]["good_with_dogs"] == True:
        dogs = "Other Dogs"
    else:
        dogs = ""
    if json_object["animal"]["home_environment_attributes"]["good_with_cats"] == True:
        cats = "Cats"
    else:
        cats = ""
    return([kids,dogs,cats])

def other(jtext):
    try:
        json_object = json.loads(jtext)
        other = json_object["animal"]["attributes"]
        return other
    except:
        return('')  

def needs(jtext):
    try:
        json_object = json.loads(jtext)
        needs = json_object["animal"]["special_needs_notes"]
        return needs
    except:
        return('')  
    
petsdf['Breed'] = petsdf['jtext'].apply(breed)
petsdf['GoodWith'] = petsdf['jtext'].apply(goodwith)
petsdf['Attributes'] = petsdf['jtext'].apply(other)
petsdf['SpecialNeeds'] = petsdf['jtext'].apply(needs)

petsdf.head()
```
